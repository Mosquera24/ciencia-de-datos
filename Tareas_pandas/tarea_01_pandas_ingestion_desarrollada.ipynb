**¿Cuántas personas tienen la palabra Chief en el título de su trabajo?** Apoyese en los métodos para manipular strings y en el método `.apply()`

```python
# Copie el código aca
```

```python
salarios["JobTitle"].apply(lambda x: "chief" in str(x).lower()).sum()
```

Este código es valido, se recomienda hacer el código mas simple y agregar typing y documentación.

1. Seleccionamos la columna [JobTitle] del DataFrame.
2. .str permite usar métodos de texto (strings) sobre toda la columna. Es lo que hace posible aplicar funciones como contains a cada fila.
3. Los argumentos dentro del paréntesis "chief" → texto que se busca
case=False → ignora mayúsculas/minúsculas
na=False → si hay valores nulos, los trata como False (evita errores)
4. .sum Suma los valores booleanos

```python
salarios["JobTitle"].str.contains("chief", case=False, na=False).sum()
```
