### Punto 4. Agregación simple

Calcula:
- número total de cursos
- precio promedio
- precio mínimo
- precio máximo

Usa la tabla `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 4 - SQL
# Escribe aquí tu consulta
run_sql('''
SELECT
    COUNT(*) AS total_cursos,
    AVG(precio) AS precio_promedio,
    MIN(precio) AS precio_minimo,
    MAX(precio) AS precio_maximo
FROM cursos;''')
```

```python
# Punto 4 - pandas
# Escribe aquí tu solución

resultado = cursos ["precio"].agg(
    total_cursos = "count",
    precio_promedio = "mean",
    precio_minimo="min",
    precio_maximo="max"
)

display(resultado)
```

### Punto 6. JOIN entre tablas

Muestra:
- nombre del estudiante
- nombre del curso
- estado de la inscripción

Usa `estudiantes`, `inscripciones` y `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 6 - SQL
# Escribe aquí tu consulta

run_sql ('''
SELECT
    e.nombre AS nombre_estudiante,
    c.nombre_curso,
    i.estado
FROM inscripciones i
JOIN estudiantes e
    ON e.estudiante_id = i.estudiante_id
JOIN cursos c
    ON c.curso_id = i.curso_id;''')
```

```python
# Punto 6 - pandas
# Escribe aquí tu solución

# Unión estudiantes + inscripciones
df_merge = inscripciones.merge(estudiantes, on="estudiante_id")

# Unión con cursos
df_final = df_merge.merge(cursos, on="curso_id")

# Seleccionar columnas
resultado = df_final[["nombre", "nombre_curso", "estado"]]

display(resultado)
```

### Punto 7. LEFT JOIN

Trae todos los estudiantes, incluso si no tuvieran inscripciones, mostrando:
- `estudiante_id`
- `nombre`
- `inscripcion_id`

**Entregable:**
- SQL
- pandas

> Pista: en este dataset puede que todos tengan al menos una inscripción, pero igual debes practicar el `LEFT JOIN`.

```python
# Punto 7 - SQL
# Escribe aquí tu consulta

run_sql ('''
SELECT
    e.estudiante_id,
    e.nombre,
    i.inscripcion_id
FROM estudiantes e
LEFT JOIN inscripciones i
    ON e.estudiante_id = i.estudiante_id;''')
```

```python
# Punto 7 - pandas
# Escribe aquí tu solución

estudiantes_total = estudiantes.merge(
    inscripciones,
    on="estudiante_id",
    how="left"
)[["estudiante_id", "nombre", "inscripcion_id"]]

display(estudiantes_total)
```

### Punto 8. HAVING

Calcula cuántas inscripciones tiene cada estudiante y muestra solo aquellos con **2 o más**.

Debes mostrar:
- `estudiante_id`
- `nombre`
- `n_inscripciones`

**Entregable:**
- SQL
- pandas

```python
# Punto 8 - SQL
# Escribe aquí tu consulta

run_sql ('''
SELECT
    e.estudiante_id,
    e.nombre,
    COUNT(i.inscripcion_id) AS n_inscripciones
FROM estudiantes e
JOIN inscripciones i
    ON e.estudiante_id = i.estudiante_id
GROUP BY e.estudiante_id, e.nombre
HAVING COUNT(i.inscripcion_id) >= 2;''')
```

```python
# Punto 8 - pandas
# Escribe aquí tu solución

inscripciones = (
    inscripciones
    .groupby("estudiante_id")
    .size()
    .reset_index(name="n_inscripciones")
    .merge(estudiantes, on="estudiante_id")
)

# Filtrar los que tienen 2 o más
final = inscripciones[inscripciones["n_inscripciones"] >= 2][
    ["estudiante_id", "nombre", "n_inscripciones"]
]

display(final)
```

### Punto 9. CASE WHEN / segmentación

Clasifica los cursos según su precio:
- `Bajo` si el precio es menor a 120
- `Medio` si el precio está entre 120 y 180 inclusive
- `Alto` si el precio es mayor a 180

Debes mostrar:
- `nombre_curso`
- `precio`
- `segmento_precio`

**Entregable:**
- SQL
- pandas

```python
# Punto 9 - SQL
# Escribe aquí tu consulta

run_sql ('''
SELECT
    nombre_curso,
    precio,
    CASE
        WHEN precio < 120 THEN 'Bajo'
        WHEN precio BETWEEN 120 AND 180 THEN 'Medio'
        ELSE 'Alto'
    END AS segmento_precio
FROM cursos;''')
```

```python
# Punto 9 - pandas
# Escribe aquí tu solución
import numpy as np

cursos_segmentados = cursos.copy()

cursos_segmentados["segmento_precio"] = np.where(
    cursos_segmentados["precio"] < 120, "Bajo",
    np.where(
        cursos_segmentados["precio"] <= 180, "Medio",
        "Alto"
    )
)

# Seleccionar columnas
resultado_cursos = cursos_segmentados[["nombre_curso", "precio", "segmento_precio"]]

display(resultado_cursos)
```

### Punto 10. Ejercicio final integrador

Calcula el total pagado por categoría de curso.

Debes:
- unir `cursos`, `inscripciones` y `pagos`
- agrupar por `categoria`
- sumar `monto`
- ordenar de mayor a menor

**Entregable:**
- SQL
- pandas

```python
# Punto 10 - SQL
# Escribe aquí tu consulta

run_sql ('''
SELECT
    c.categoria,
    SUM(p.monto) AS total_pagado
FROM pagos p
JOIN inscripciones i
    ON p.inscripcion_id = i.inscripcion_id
JOIN cursos c
    ON i.curso_id = c.curso_id
GROUP BY c.categoria
ORDER BY total_pagado DESC;''')
```

```python
# Punto 10 - pandas
# Escribe aquí tu solución

pagos_cursos = (
    pagos
    .merge(inscripciones, on= "inscripcion_id")
    .merge(cursos, on="curso_id")
    .groupby("categoria")["monto"]
    .sum()
    .sort_values(ascending=False)
    .reset_index(name="total_pagado")
)

display(pagos_cursos)
```
